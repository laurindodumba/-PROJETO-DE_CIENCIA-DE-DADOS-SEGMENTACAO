#### Amostra balanceada

```python
# Supondo que df seja o seu dataframe e 'target' a sua coluna de interesse
evento = abt_00[abt_00['target'] == 1]
nevento = abt_00[abt_00['target'] == 0]

# Amostragem aleatria dos negativos
negativos_amostra = nevento.sample(n=len(evento))

# Concatenar as duas partes para obter a amostra balanceada
amostra_balanceada = pd.concat([evento, negativos_amostra])

# Embaralhar os dados, se desejado
amostra_balanceada = amostra_balanceada.sample(frac=1).reset_index(drop=True)

amostra_balanceada.shape
```

```python
amostra_balanceada.target.describe()
```

```python
# Separando as variveis de entrada (features) e de sada (target)
X = amostra_balanceada.drop(columns=["target"])
y = amostra_balanceada["target"]
```

```python
# Separando uma amostra de 70% para treinar o modelo e 30% para testar o modelo
# Holdout 70/30 (out-of-sample)

from sklearn.model_selection import train_test_split

# Dividindo os dados em conjunto de treinamento e teste (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Definindo o modelo
model = DecisionTreeClassifier(random_state=42)

# Definindo os parmetros para o grid search
param_grid = {
    'criterion':['gini','entropy'],
    'splitter':['best'],
    'max_depth': [10,20,30], # Profundidade mxima da rvore
    'min_samples_split': [0.008, 0.01], # Nmero mnimo de amostras necessrias para dividir um n interno.
    'min_samples_leaf': [0.02,0.04], # Nmero mnimo de amostras necessrias para estar em um n folha.
}

# Calculando a quantidade total de modelos que sero treinados
num_models = len(param_grid['splitter']) * len(param_grid['criterion']) * len(param_grid['max_depth']) * len(param_grid['min_samples_split']) * len(param_grid['min_samples_leaf']) * 5  # 5  o nmero de folds na validao cruzada (cv)
print(f"Total de Modelos a serem Treinados: {num_models}")

# Definindo o objeto GridSearchCV
grid = GridSearchCV(model,
                    param_grid,
                    cv=5,
                    scoring='roc_auc',  # 'accuracy' 'precision' 'recall' 'f1'
                    n_jobs=-1)

# Treinando o modelo com o grid search
grid.fit(X_train, y_train)

# Exibindo os melhores parmetros encontrados pelo grid search
print("Melhores Parmetros: ", grid.best_params_)

# Exibindo a melhor pontuao (score) atingida pelo modelo com os melhores parmetros
print("Melhor AUC: ", grid.best_score_)

# Utilizando o melhor modelo para fazer previses
predictions = grid.best_estimator_.predict(X_test)
predprob = grid.best_estimator_.predict_proba(X_test)

gini_n = round(pod.gini_normalizado(y_test, predprob[:, 1]),2)
print('Gini normalizado --> ',gini_n)

best_model_dt = grid.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Definindo o modelo
model = RandomForestClassifier()

# n_estimators=100,  # Nmero de rvores na floresta.
# criterion='gini',  # Funo para medir a qualidade de uma diviso ("gini" ou "entropy").
# max_depth=None,  # Profundidade mxima das rvores. None significa que os ns so expandidos at ficarem puros.
# min_samples_split=2,  # Nmero mnimo de amostras necessrias para dividir um n interno.
# min_samples_leaf=1,  # Nmero mnimo de amostras necessrias para estar em um n folha.
# min_weight_fraction_leaf=0.0,  # Frao mnima ponderada da soma total de pesos necessria para estar em um n folha.
# max_features='auto',  # Nmero de caractersticas a considerar ao procurar a melhor diviso.
# max_leaf_nodes=None,  # Nmero mximo de ns folha.
# min_impurity_decrease=0.0,  # Um n ser dividido se a diviso induzir a uma diminuio da impureza maior ou igual a este valor.
# bootstrap=True,  # Se as amostras de bootstrap so usadas ao construir rvores.
# oob_score=False,  # Se usar amostras out-of-bag para estimar a preciso generalizada.
# n_jobs=None,  # Nmero de jobs a serem executados em paralelo para ajuste e previso. -1 significa usar todos os processadores.
# random_state=None,  # Controla a aleatoriedade do bootstrap e da seleo de features.
# verbose=0,  # Controla a verbosidade do processo de treinamento.
# warm_start=False,  # Reutiliza a soluo da chamada anterior para ajustar e adicionar mais estimadores ao ensemble.
# class_weight=None  # Pesos associados s classes. Pode ser "balanced", "balanced_subsample" ou um dicionrio com os pesos.


# Definindo os parmetros para o grid search
param_grid = {
    'n_estimators': [10, 50,100],
    'max_depth': [2, 3, 5],
    'min_samples_split': [0.001, 0.01],
    'min_samples_leaf': [2, 0.01],
    'bootstrap': [True, False],
    # 'max_features': ['auto', 'sqrt', 'log2']
}

# Definindo o objeto GridSearchCV
grid = GridSearchCV(model,
                    param_grid,
                    cv=5,
                    verbose=2,
                    scoring='roc_auc',  # 'accuracy' 'precision' 'recall' 'f1'
                    n_jobs=-1)

# Treinando o modelo com o grid search
grid.fit(X_train, y_train)

# Exibindo os melhores parmetros encontrados pelo grid search
print("Melhores Parmetros: ", grid.best_params_)

# Exibindo a melhor pontuao (score) atingida pelo modelo com os melhores parmetros
print("Melhor AUC: ", grid.best_score_)

# Utilizando o melhor modelo para fazer previses
predictions = grid.best_estimator_.predict(X_test)

predprob = grid.best_estimator_.predict_proba(X_test)

gini_n = round(pod.gini_normalizado(y_test, predprob[:, 1]),2)
print('Gini normalizado --> ',gini_n)

best_model_rf = grid.best_estimator_
```
